```r
data=iris[,1:4]
label=iris[,5]
```

```r
iris.pca <- princomp(data ,center = TRUE, scale = TRUE)  # make sure to normalize your data before PCA
```

```r
summary(iris.pca)
```

```r
plot(iris.pca, type = "l")
```

```r
#library(devtools)
#install_github("vqv/ggbiplot")
 
library(ggbiplot)
g <- ggbiplot(iris.pca, obs.scale = 1, var.scale = 1, 
              groups = label, ellipse = TRUE, 
              circle = TRUE)
g <- g + scale_color_discrete(name = '')
g <- g + theme(legend.direction = 'horizontal', 
               legend.position = 'top')
print(g)
```
