```r
library(ggplot2)
```

```r
id=1:16
weight=sample(50:100,16)
lenght=sample(90:190,16)
sex=sample(c("male","femal"),16,replace=T)
ethnicity=sample(c("asian","african americain","indian"),16,replace=T)
tumer_size=runif(16,1,3)
drug=sample(c('drug A','drug B'),16,replace=T)
live=sample(c('live','dead'),16,replace=T)
data=data.frame(id,weight,lenght,sex,ethnicity,tumer_size,drug,live)
head(data)
tail(data)
```

```r
library(ggplot2)
ggplot(data, aes(x=lenght,y=weight)) + geom_point()+geom_smooth(method="lm")
```

```r
library(ggplot2)
ggplot(data, aes(x=lenght,y=weight)) + geom_point(col="steelblue", size=3)+geom_smooth(method="lm")+
ggtitle("Weight-light scater plot") + xlab("Length") + ylab("Weight")
```

```r
ggplot(data, aes(x = ethnicity, y = weight)) +
  geom_boxplot(fill = "grey80", colour = "blue") +
  scale_x_discrete() + xlab("Ethnicity") +
  ylab("Weight")+theme(text = element_text(size=20),
        axis.text.x = element_text(angle=45, hjust=1)) 
```
